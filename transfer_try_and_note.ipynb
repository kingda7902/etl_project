````python
import json
import re
from os import walk
from collections import Counter


#return a dict key:path value:filenames
def getFilesPath(path = './'):
    pathdict = {}
    for (dirpath, dirnames, filenames) in walk(path):
        for name in dirnames:
            nowpath = dirpath + '/'+ name
            pathlist.update(getFilesPath(nowpath))
        pathdict[dirpath] = filenames
        break
    return pathdict

def str2list(skills):
    skills_list = re.findall('([\x00-\x7F]+)', skills)

    return [word.strip().upper() for word in skills_list]


def list_for_all(jobs):
    count_list = []
    for work in jobs:
        count_list.extend( str2list(work['tools']))
    return count_list

def counting_tools(path):
    pathDict = getFilesPath(path)
    ct = Counter()
    ct['jobs_number'] = 0
    for key, value in pathDict.items():
        for filename in value:
            if '.json' in filename:
                filepath = key+'/'+filename
                with open(filepath, encoding = 'utf8') as f:
                    datas = json.load(f)
                    count_list = list_for_all(datas['job_lists'])
                    for word in count_list:
                        ct[word] += 1
                        ct['jobs_number'] += 1
    return ct


if __name__ == "__main__":
    path = './data/104'
    ct = counting_tools(path)
````

````python
import requests
import time
apiurl = 'https://maps.googleapis.com/maps/api/geocode/json'

def pregeocoding(latlng, key=''):
    queryData = {}
    queryData['latlng'] = latlng
    queryData['key'] = key
    queryData['language'] = 'zh-TW'
    res = requests.get(apiurl, params=queryData)
    geo = res.json()
    time.sleep(0.02)
    res.close()
    if geo.get('status') == 'OK':       
        formatted_address=list(map( lambda x : x.get('formatted_address') , geo['results']))
        address_list = list(filter( lambda x : re.match(r'^\d+.*[號]',x), formatted_address))
        if len(address_list) > 0:
            return {'address':address_list[0]}
        else:
            return {'address':max(formatted_address, key=len)}
    else:
        return {'status':geo.get('status')}

    
````
