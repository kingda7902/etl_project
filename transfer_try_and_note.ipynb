## read data from a folder

`os.walk` return tuple (dirpath, dirnames, filenames)

```python
#return a dict key:path value:filenames
from os import walk

def getFilesPath(path = './'):
    pathlist = {}
    for (dirpath, dirnames, filenames) in walk(path):
        for name in dirnames:
            nowpath = dirpath + '/'+ name
            pathlist.update(getFilesPath(nowpath))
        pathlist[dirpath] = filenames
        break
    return pathlist
```

```python
import json
import re
from os import walk
from collections import Counter


#return a dict key:path value:filenames
def getFilesPath(path = './'):
    pathdict = {}
    for (dirpath, dirnames, filenames) in walk(path):
        for name in dirnames:
            nowpath = dirpath + '/'+ name
            pathlist.update(getFilesPath(nowpath))
        pathdict[dirpath] = filenames
        break
    return pathdict

def str2list(skills):
    skills_list = re.findall('([\x00-\x7F]+)', skills)

    return [word.strip().upper() for word in skills_list]


def list_for_all(jobs):
    count_list = []
    for work in jobs:
        count_list.extend( str2list(work['tools']))
    return count_list

def counting_tools(path):
    pathDict = getFilesPath(path)
    ct = Counter()
    ct['jobs_number'] = 0
    for key, value in pathDict.items():
        for filename in value:
            if '.json' in filename:
                filepath = key+'/'+filename
                with open(filepath, encoding = 'utf8') as f:
                    datas = json.load(f)
                    count_list = list_for_all(datas['job_lists'])
                    for word in count_list:
                        ct[word] += 1
                        ct['jobs_number'] += 1
    return ct


if __name__ == "__main__":
    path = './data/104'
    ct = counting_tools(path)
```

```python
ct.most_common(31)
```

```python
import re
import json

def loads_jsonp(_jsonp):
    try:
        return json.loads(re.match(".*?({.*}).*",_jsonp,re.S).group(1))
    except:
        raise ValueError('Invalid Input')
```
